# Univariate , Bivariate , Multivariate  Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib.inline
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.shape
```

```python
df
```

### Univariate

```python
df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'*')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'*')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'*')
plt.xlabel('sepal_length')
```

### Bivariate

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,'sepal_length','sepal_width').add_legend();
```

```python
sns.FacetGrid(df,hue='species',size=8).map(plt.scatter,'petal_length','petal_width').add_legend();
```

### Multivariate 

```python
sns.pairplot(df,hue='species',size=3)
```

```python
sns.pairplot(df,hue='sepal_width',size=4)
```

```python
sns.pairplot(df,hue='sepal_length',size=4)
```

```python
sns.pairplot(df,hue='petal_length',size=4)
```

```python
sns.pairplot(df,hue='petal_width',size=3)
```
